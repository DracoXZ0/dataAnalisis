## Punto 1

Cargue el archivo home_data.csv e identifique cuantas variables y registros tiene.

## Punto 2

Haga un histograma de la variable 'price' y calcule su kurtosis y asimetría. ¿Parece ser una distribución normal?

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histograma de la variable "price"')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

kurtosis = stats.kurtosis(df['price'])
asimetria = stats.skew(df['price'])

print("Kurtosis:", kurtosis)
print("Asimetría:", asimetria)
print("No, la distribucion no corresponde a una normal, sino una campana a la izquierda")
```

##  Punto 3

Haga un boxplot de la variable 'price'. ¿Qué se puede deducir de este gráfico?

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'], vert=False)
plt.title('Boxplot de la variable "price"')
plt.xlabel('Precio')
plt.yticks([])
plt.grid(True)
plt.show()
print("Se puede deducir que existe una cantidad considerable de outliers")
```

## Punto 4

Eliminar los registros que de acuerdo con el boxplot son outliers de la variable 'price'.

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

outliers_removed = len(df) - len(df_filtered)
print("Registros eliminados:", outliers_removed)
```

```python
print(len(df_filtered))
```

## Punto 5

Vuelva a graficar el histograma de la variable 'price' y a calcular su kurtosis y asimetría, pero esta vez en el dataframe sin los outliers. ¿Parece ser una distribución normal?

```python
plt.figure(figsize=(10, 6))
plt.hist(df_filtered['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histograma de la variable "price"')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

kurtosis = stats.kurtosis(df_filtered['price'])
asimetria = stats.skew(df_filtered['price'])

print("Kurtosis:", kurtosis)
print("Asimetría:", asimetria)
print("Se acerca a una distribucion normal, pero sesgada hacia la izquierda")
```

## Punto 6

## Punto 7

Ahora calcule la media, la mediana y la desviacion estándar de la variable 'price' agrupada por la variable 'condition', en el dataframe sin outliers.
- ¿Existen diferencias apreciables en estas estadísticas entre grupos?
- ¿En que condición la media del precio de las casas es mayor?

## Punto 13

Genere una nueva variable que se va a llamar 'ratio_price_area_by_condition' que va a ser igual a:

$$ ratio\_price\_area\_by\_condition = \frac{price - mean(price_{condition})}{sqft\_living - mean(sqft\_living_{condition})}
$$

Donde $mean(price_{condition})$ es la media de la variable 'price' calculada para la condición a la que pertenece la muestra, y $mean(sqft\_living_{condition})$ es la media de la variable 'sqft_living' calculada para la condición a la que pertenece la muestra.

```python

mean_price_by_condition = df_filtered.groupby('condition')['price'].transform('mean')
mean_sqft_living_by_condition = df_filtered.groupby('condition')['sqft_living'].transform('mean')

df_filtered['ratio_price_area_by_condition'] = (df_filtered['price'] - mean_price_by_condition) / (df_filtered['sqft_living'] - mean_sqft_living_by_condition)

df_filtered
```

## Punto 14

Cree un nuevo dataframe filtrando los registros del  dataframe original que pertenezcan a valores (categorías) de las variable 'grade' que tengan más de 1000 muestras.

- ¿Cuántos registros quedan?
- ¿Que valores puede tomar la variable 'grade' en el nuevo dataframe?

```python
grade_counts = df_filtered['grade'].value_counts()

grades_mas_de_1000 = grade_counts[grade_counts > 1000].index.tolist()

nuevo_df = df_filtered[df_filtered['grade'].isin(grades_mas_de_1000)]

print("Cantidad de registros en el nuevo DataFrame:", len(nuevo_df))

print("Valores de 'grade' en el nuevo DataFrame:", nuevo_df['grade'].unique())
```
