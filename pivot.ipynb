```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Va a trabajar con 3 conjuntos de datos, uno conteniendo las abreviaturas de los diferentes estados de Estados Unidos, otro las áreas de estos estados, y el tercero conteniendo datos de la población y la población joven de estos estados en diferentes años.

Cargue los 3 datasets, cada uno en un dataframe.

```python
df = pd.read_csv('state-abbrevs.csv')
df
```

```python
df1 = pd.read_csv('state-areas.csv')
df1
```

```python
combined_df1 = pd.merge(df, df1, on='state')
combined_df1
```

```python
df2 = pd.read_csv('state-population.csv')
df2
```

```python
merged_df = pd.merge(combined_df1, df2, left_on='abbreviation', right_on='state/region', how='inner')

merged_df.drop('state/region', axis=1, inplace=True)

merged_df
```

Seleccione los 10 estados con mayor población en el año 2013. No incluya la población total del país.

Haga un gráfico de líneas mostrando la población de cada uno de estos estados en los años cuya información contiene el dataset de población. En el dataset debe aparecer el nombre del estado, no la abreviatura.


```python
df2013 = merged_df[(merged_df.year == 2013) & (merged_df.ages != 'total')]
df2013 = df2013.sort_values(by='population', ascending=False).head(10)
df2013
```

```python
merged_df['year'] = pd.to_numeric(merged_df['year'])
merged_df['population'] = pd.to_numeric(merged_df['population'])

states_2013 = df2013['state'].unique()
filtered_df = merged_df[merged_df['state'].isin(states_2013)]

grouped_df = filtered_df.groupby(['state', 'year'])['population'].sum().reset_index()

for state in states_2013:
    state_data = grouped_df[grouped_df['state'] == state]
    plt.plot(state_data['year'], state_data['population'], label=state)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth Over Time')
plt.legend()
plt.show()
```

¿Cuál es el estado en el que más ha crecido la población en este periodo de tiempo?

¿Cuál es el estado en el que porcentualmente más ha crecido la población en este periodo de tiempo?

```python
grouped_df['year'] = pd.to_numeric(grouped_df['year'])
grouped_df['population'] = pd.to_numeric(grouped_df['population'])

pop_initial = grouped_df.groupby('state').apply(lambda x: x.loc[x['year'].idxmin()]['population'])
pop_final = grouped_df.groupby('state').apply(lambda x: x.loc[x['year'].idxmax()]['population'])
pop_growth = pop_final - pop_initial

max_growth_state = pop_growth.idxmax()
max_growth_state
```

```python
pop_growth_percentage = (pop_final - pop_initial) / pop_initial * 100

max_growth_percentage_state = pop_growth_percentage.idxmax()

max_growth_percentage_state
```

Ahora haga lo mismo, pero con la población menor a 18 años.

```python
df_menor = merged_df[merged_df['ages'] == 'under18']

poblacion_inicial_menor_18 = df_menor.groupby('state').apply(lambda x: x.loc[x['year'].idxmin()]['population'])
poblacion_final_menor_18 = df_menor.groupby('state').apply(lambda x: x.loc[x['year'].idxmax()]['population'])

crecimiento_m = poblacion_final_menor_18 - poblacion_inicial_menor_18
```

```python
max_growth_state_menor_18 = crecimiento_m.idxmax()

porcentaje_18 = (poblacion_final_menor_18 - poblacion_inicial_menor_18) / poblacion_inicial_menor_18 * 100

max_growth_percentage_state_menor_18 = porcentaje_18.idxmax()
```

¿Cuál es el estado en el que más ha crecido la población menor a 18 años en este periodo de tiempo?

¿Cuál es el estado en el que porcentualmente más ha crecido la población menor a 18 años en este periodo de tiempo?

```python
print("Estado con mayor crecimiento de población menor a 18 años:", max_growth_state_menor_18)
print("Estado con mayor crecimiento porcentual de población menor a 18 años:", max_growth_percentage_state_menor_18)
```

¿Cuáles son los 10 estados más densamente poblados en el año 2013?

```python
merged_df['population'] = pd.to_numeric(merged_df['population'])
merged_df['area (sq. mi)'] = pd.to_numeric(merged_df['area (sq. mi)'])
merged_df['density'] = merged_df['population'] / merged_df['area (sq. mi)']
```

```python
df_2013 = merged_df[merged_df['year'] == 2013]

top_10_densely_populated_states = df_2013.nlargest(10, 'density')
```

```python
print("Los 10 estados más densamente poblados en el año 2013 son:")
print(top_10_densely_populated_states[['state', 'density']])
```

Haga una gráfica de la población total por años y de la poblaciòn menor a 18 años por años.

¿Cuáles son los 5 años en los que la proporción de población menor a 18 años es más bajo?


```python
merged_df_nousa = merged_df[merged_df['state'] != 'USA']
merged_df_nousa
```

```python
total_population_by_year = merged_df.groupby('year')['population'].sum()
plt.figure(figsize=(10, 4))
plt.plot(total_population_by_year.index, total_population_by_year.values, color='blue')
plt.title('Población Total por Años')
plt.xlabel('Año')
plt.ylabel('Población Total')
plt.grid(True)
plt.show()
```

```python
population_under_18_by_year = merged_df[merged_df['ages'] == 'under18'].groupby('year')['population'].sum()
plt.figure(figsize=(10, 4))
plt.plot(population_under_18_by_year.index, population_under_18_by_year.values, color='blue')
plt.title('Población Menor a 18 años por Años')
plt.xlabel('Año')
plt.ylabel('Población Menor a 18 años')
plt.grid(True)
plt.show()
```

```python
merged_df['population'] = pd.to_numeric(merged_df['population'])
population_under_18_by_year = merged_df[merged_df['ages'] == 'under18'].groupby('year')['population'].sum()
total_population_by_year = merged_df.groupby('year')['population'].sum()
proportion_under_18_by_year = population_under_18_by_year / total_population_by_year

lowest_proportion_years = proportion_under_18_by_year.nsmallest(5)
```

```python
print("Los 5 años con las proporciones más bajas de población menor a 18 años son:")
print(lowest_proportion_years)
```
