## 1
Cree un DataFrame con la siguiente información de al menos 5 familiares o amigos: documento de identidad, nombres, apellidos, edad, ciudad de nacimiento, país de nacimiento, ciudad de residencia, país de residencia, ocupación o profesión, años de escolaridad.

Configure el campo "documento de identidad" como índice.

```python
data = {
    'Documento de Identidad': ['123456789', '987654321', '456789123', '789123456', '321654987'],
    'Nombres': ['Juan', 'María', 'Pedro', 'Ana', 'Luis'],
    'Apellidos': ['Gómez', 'López', 'Martínez', 'Rodríguez', 'Sánchez'],
    'Edad': [30, 25, 35, 28, 40],
    'Ciudad de Nacimiento': ['Bogotá', 'Madrid', 'México DF', 'Buenos Aires', 'Lima'],
    'País de Nacimiento': ['Colombia', 'España', 'México', 'Argentina', 'Perú'],
    'Ciudad de Residencia': ['Medellín', 'Barcelona', 'Ciudad de México', 'Buenos Aires', 'Lima'],
    'País de Residencia': ['Colombia', 'España', 'México', 'Argentina', 'Perú'],
    'Ocupación o Profesión': ['Ingeniero', 'Médico', 'Abogado', 'Profesor', 'Empresario'],
    'Años de Escolaridad': [16, 18, 20, 15, 14]
}

df = pd.DataFrame(data)

df.set_index('Documento de Identidad', inplace=True)

df
```

## 2
Genere un arreglo aleatorio de numpy de tamaño 10 x 8, que tenga distribución normal con media 0 y desviación estándar 1, y con este cree un DataFrame.

```python
arreglo_aleatorio = np.random.normal(loc=0, scale=1, size=(10, 8))
df = pd.DataFrame(arreglo_aleatorio)

df
```

## 3
Ahora genere 3 columnas nuevas que contengan la suma, el promedio y la desviación estándar de los elementos de cada fila.


```python
df['Suma'] = df.sum(axis=1)
df['Promedio'] = df.mean(axis=1)
df['Desviación estándar'] = df.std(axis=1)

df
```

## 4
Por último, filtre las filas cuya media es mayor a 0, y genere un nuevo DataFrame que contenga la suma, el promedio y la desviación estándar de los elementos de cada columna.


```python
df['Promedio'] = df.mean(axis=1)

df_filtrado = df.loc[df['Promedio'] > 0]

suma_columnas = df_filtrado.sum()
promedio_columnas = df_filtrado.mean()
desviacion_std_columnas = df_filtrado.std()

nuevo_df = pd.DataFrame({
    'Suma': suma_columnas,
    'Promedio': promedio_columnas,
    'Desviación estándar': desviacion_std_columnas
})

nuevo_df
```
