```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mcu_box_office.csv')
df
```

```python
variables = df.shape[1]
observaciones = df.shape[0]
print("Número de variables:", variables)
print("Número de observaciones:", observaciones)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format="%m/%d/%Y")
```

```python
df
```

```python
df['production_budget'] = df['production_budget'].replace(to_replace=',', value='', regex=True)
df['opening_weekend'] = df['opening_weekend'].replace(to_replace=',', value='', regex=True)
df['domestic_box_office'] = df['domestic_box_office'].replace(to_replace=',', value='', regex=True)
df['worldwide_box_office'] = df['worldwide_box_office'].replace(to_replace=',', value='', regex=True)
```

```python
df
```

```python
df['production_budget'] = df['production_budget'].astype('int64')
df['opening_weekend'] = df['opening_weekend'].astype('int64')
df['domestic_box_office'] = df['domestic_box_office'].astype('int64')
df['worldwide_box_office'] = df['worldwide_box_office'].astype('int64')
```

```python
df
```

```python


nominales = ['movie_title']
ordinales = ['release_date', 'mcu_phase' ]
numericas = ['mcu_phase', 'tomato_meter', 'audience_score', 'movie_duration', 'production_budget', 'opening_weekend', 'domestic_box_office', 'worldwide_box_office']
```

```python
df.set_index('movie_title', inplace=True)

print("Variables nominales:", nominales)
print("Variables ordinales:", ordinales)
print("Variables numéricas:", numericas)
```

13. Cree una nueva columna llamada 'ganancias', que se calcula restando 'production_budget' de 'worldwide_box_office'.
14. Ordene las fases del MCU de acuerdo al promedio de ganancias de las películas que pertenecen a cada fase.
15. Indique cual fue la película que produjo más ganancias en cada fase, y cual fue la ganancia que produjo esta película.
Tip: en el último punto puede usar el método idxmax().

```python
df = df.assign(ganancias = df['worldwide_box_office'] - df['production_budget'])
df
```

```python
promedio_ganancias_por_fase = df.groupby('mcu_phase')['ganancias'].mean()

fases_ordenadas = promedio_ganancias_por_fase.sort_values(ascending=False)

print("Fases del MCU ordenadas según el promedio de ganancias:")
fases_ordenadas
```

```python
ganancia_max_por_fase = df.groupby('mcu_phase')['ganancias'].idxmax()
for fase, indice in ganancia_max_por_fase.items():
    ganancia_max = df.loc[indice, 'ganancias']
    print(f"Fase {fase}: La película que produjo más ganancias fue '{indice}' con una ganancia de {ganancia_max:,} dólares.")
```

16. Haga gráficos de dispersión de 'ganancias' versus cada una de estas variables: 'tomato_meter', 'audience_score', 'movie_duration', 'production_budget', 'opening_weekend', 'domestic_box_office', 'worldwide_box_office'.
17. De acuerdo con estas gráficas, ¿con qué variables tiene relación más fuerte la ganancia?
18. Ahora grafique el comportamiento de las ganancias en el tiempo, usando el método plot.line() y poniendo en x 'release_date'. ¿Se observa alguna tendencia en el tiempo?

```python
dispersion = ['tomato_meter', 'audience_score', 'movie_duration', 'production_budget', 'opening_weekend', 'domestic_box_office', 'worldwide_box_office']

plt.figure(figsize=(15, 10))
for i, variable in enumerate(dispersion, 1):
    plt.subplot(3, 3, i)
    plt.scatter(df[variable], df['ganancias'], alpha=0.5, color='skyblue')
    plt.title(f'Ganancias vs {variable}')
    plt.xlabel(variable)
    plt.ylabel('Ganancias')
plt.tight_layout()
plt.show()
```

### Como podemos obvservar, las variables que mas estan relacionadas con ganancias son worldwide_box_office, opening_weekend y domestic_box_office

### Por que podemos asumir eso, si nosotros trazamos una recta, esta pasa por la mayoria de los puntos y con una pendiente cercana a 1, eso significa que hay una relacion estrecha

```python
df_sorted_by_date = df.sort_values('release_date')

plt.figure(figsize=(15, 6))
plt.plot(df_sorted_by_date['release_date'], df_sorted_by_date['ganancias'], color='skyblue')
plt.title('Ganancias en el tiempo')
plt.xlabel('Fecha de lanzamiento')
plt.ylabel('Ganancias')
plt.grid(True)
plt.show()
```

### Se podria decir, que hay una tendencia, cual? fluctuante, si vemos la grafica, hay unos picos importantes entre 2018 y 2019, vemos tambien que hay una caida particular en el 2020, a que se debe esto, al covid, el recaudo de taquillas pues fue menor, se recupera en el 2022 cuando se vuelve a la "normalidad"
